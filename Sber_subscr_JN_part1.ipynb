```python
def print_useful_rows_info(df):
    print('Количество полностью заполненных объектов из всей выборки:', 
          len(df.dropna()))
    print('Процент полностью заполненных объектов из всей выборки:', 
          round(len(df.dropna()) / len(df) * 100, 2))
    
print_useful_rows_info(df_ga_hits)
print_useful_rows_info(df_ga_sessions)
```

```python
missing_values_ga_hits = ((df_ga_hits.isna().sum() / len(df_ga_hits)) * 100).sort_values(ascending=False)
missing_values_ga_sessions = ((df_ga_sessions.isna().sum() / len(df_ga_sessions)) * 100).sort_values(ascending=False)
print('Процент пропущенных значений:\n', missing_values_ga_hits)
print('Процент пропущенных значений:\n', missing_values_ga_sessions)
```

```python
df_ga_hits_clean = df_ga_hits.copy()
df_ga_sessions_clean = df_ga_sessions.copy()
```

```python
# Удаляем пустые колонки или колонки c пропусками больше 20%
df_ga_hits_clean = df_ga_hits_clean.drop(columns=['event_value', 'hit_time', 'hit_referer', 'event_label'])
df_ga_sessions_clean = df_ga_sessions_clean.drop(columns=['device_model', 'utm_keyword', 'device_os', 'device_brand'])
```

```python
# Проверка наличия дубликатов
print(df_ga_hits_clean.duplicated().sum())
print(df_ga_sessions_clean.duplicated().sum())
```

```python
df_ga_hits_clean = df_ga_hits_clean.drop_duplicates()
```

```python
# Проверяем повторно дубликаты
print(df_ga_hits_clean.duplicated().sum())
```

```python
missing_values_ga_hits_clean = ((df_ga_hits_clean.isna().sum() / len(df_ga_hits_clean)) * 100).sort_values(ascending=False)
missing_values_ga_sessions_clean = ((df_ga_sessions_clean.isna().sum() / len(df_ga_sessions_clean)) * 100).sort_values(ascending=False)
print('Процент пропущенных значений:\n', missing_values_ga_hits_clean)
print('Процент пропущенных значений:\n', missing_values_ga_sessions_clean)
```

```python
merged_df = pd.merge(df_ga_hits_clean, df_ga_sessions_clean, on='session_id', how='outer')
```

```python
merged_df.tail(100)
```

```python
merged_df_cp = merged_df.copy()
```

```python
print_useful_rows_info(merged_df_cp)
```

```python
missing_values_merged_df_cp = ((merged_df_cp.isna().sum() / len(merged_df_cp)) * 100).sort_values(ascending=False)
print('Процент пропущенных значений:\n', missing_values_merged_df_cp)
```

```python
msno.bar(merged_df_cp)
```

```python
print(merged_df_cp.duplicated().sum())
```

```python
# обработаем пропуски
merged_df_cp['utm_adcontent'].fillna('other', inplace=True)
merged_df_cp['utm_campaign'].fillna('other', inplace=True)
merged_df_cp.dropna(subset=['hit_number', 'hit_type', 'hit_page_path', 'event_category', 'event_action', 
                  'hit_date', 'utm_source', 'visit_number', 'geo_country', 'device_browser', 
                  'device_screen_resolution', 'device_category', 'geo_city', 'utm_medium', 
                  'visit_time', 'visit_date', 'client_id'], inplace=True)
```

```python
missing_values_merged_df_cp = ((merged_df_cp.isna().sum() / len(merged_df_cp)) * 100).sort_values(ascending=False)
print('Процент пропущенных значений:\n', missing_values_merged_df_cp)
```

```python
# Удаляем колонку с одинаковыми значениями
merged_df_cp = merged_df_cp.drop(columns=['hit_type'])
```

```python
# Удаляем колонку session_id
merged_df_cp = merged_df_cp.drop(columns=['session_id'])
```

```python
cols = list(merged_df_cp.columns)
cols.insert(0, cols.pop(cols.index('client_id')))
merged_df_cp = merged_df_cp.loc[:, cols]
```

```python
merged_df_cp.head(50)
```

```python
# Удаляем колонку hit_date
merged_df_cp = merged_df_cp.drop(columns=['hit_date'])
```

```python
merged_df_cp.info()
```

```python
# преобразуем колонки в целый формат
merged_df_cp[['hit_number', 'visit_number']] = merged_df_cp[['hit_number', 'visit_number']].astype(int)

# преобразуем колонки в формат даты и времени
merged_df_cp['visit_datetime'] = merged_df_cp['visit_date'] + merged_df_cp['visit_time']
```

```python
merged_df_cp['visit_datetime'] = pd.to_datetime(merged_df_cp['visit_datetime'], format='%Y-%m-%d%H:%M:%S')
```
